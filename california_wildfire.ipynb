```python
# Dependencies
import pandas as pd
from datetime import datetime
```

```python
# Reading csv file
wildfire_data = pd.read_csv("./data_resource/california_wildfire_data_cleaned.csv")
```

```python
# Sample data from the csv file
wildfire_data.head(5)
```

```python
# Checking for the datatype of column "Fire Started"
wildfire_data["Fire Started"]
     
```

```python
# Converting the column to datetime format
wildfire_data["Fire Started"] = pd.to_datetime(wildfire_data["Fire Started"])
wildfire_data["Fire Started"]
```

```python
# checking for the datatype of column "Fire Extinguished"
wildfire_data["Fire Extinguished"]
```

```python
# Converting the column to datetime format and checking for errors
wildfire_data["Fire Extinguished"] = pd.to_datetime(wildfire_data["Fire Extinguished"], errors='coerce')
wildfire_data["Fire Extinguished"]
```

```python
# Droping rows with Not a Time value
wildfire_data.dropna(subset = ["Fire Extinguished"], inplace = True)
wildfire_data["Fire Extinguished"]
```

```python
# Collecting years from Fire Started column and creating a new column to store the details
wildfire_data["Fire Started year"] = pd.DatetimeIndex(wildfire_data["Fire Started"]).year
wildfire_data.head(5)
```

```python
# Collecting years from Fire Extinguished column and creating a new column to store the details
wildfire_data["Fire Extinguished year"] = pd.DatetimeIndex(wildfire_data["Fire Extinguished"]).year
wildfire_data.head(5)
```

```python
# Finding the dataset rows with different starting and ending year

# Creating an empty list to store the results
different_years = []
same_years = []

for i in range(len(wildfire_data)):
    if wildfire_data.iloc[i]["Fire Started year"] != wildfire_data.iloc[i]["Fire Extinguished year"]:
        different_years.append(wildfire_data.iloc[i])

    elif wildfire_data.iloc[i]["Fire Started year"] == wildfire_data.iloc[i]["Fire Extinguished year"]:
        same_years.append(wildfire_data.iloc[i])

# Creating DataFrames from the lists
different_years_df = pd.DataFrame(different_years)    # if the data falls in different years
same_years_df = pd.DataFrame(same_years)        # if the data falls in same year
same_years_df
```

## Calculations based on the same_years_df

```python
# Calculating total acres burned per year
acres_burned_per_year = same_years_df.groupby("Fire Started year")["Acres Burned"].sum().reset_index()
acres_burned_per_year.columns = ["Year", "Total Acres Burned"]
acres_burned_per_year
```

```python
# Calculating total counties burned per year
counties_burned_per_year = same_years_df.groupby("Fire Started year")["County"].nunique().reset_index()
counties_burned_per_year.columns = ["Year","Number of Counties Burned"]
counties_burned_per_year
```

```python
# Calculating total personnel involved per year
personnel_involved_per_year = same_years_df.groupby("Fire Started year")["Personnel Involved"].sum().reset_index()
personnel_involved_per_year.columns = ["Year","Personnel Involved"]
personnel_involved_per_year
```

```python
# Calculating total Structures Damaged per year
structures_damaged_per_year = same_years_df.groupby("Fire Started year")["Structures Damaged"].sum().reset_index()
structures_damaged_per_year.columns = ["Year","Structures Damaged"]
structures_damaged_per_year
```

```python
# Calculating total Structures Destroyed per year
structures_destroyed_per_year = same_years_df.groupby("Fire Started year")["Structures Destroyed"].sum().reset_index()
structures_destroyed_per_year.columns = ["Year","Structures Destroyed"]
structures_destroyed_per_year
```

## Calculations based on different_years_df

```python
# Create a list to hold the number of active days until year end
active_days_until_year_end = []

different_years_df["Fire Started"] = different_years_df["Fire Started"].dt.tz_localize(None)

for i in range(len(different_years_df)):

    start_date = different_years_df.iloc[i]["Fire Started"]
    
    year_end = datetime(start_date.year,12,31)

    active_days = (year_end - start_date).days + 1 
    active_days_until_year_end.append(active_days)

different_years_df["Active Days Until Year End"] = active_days_until_year_end
different_years_df["Remaining Days from the begining"] = different_years_df["days_of_wildfire"] - active_days_until_year_end

different_years_df
        
```

```python

# Calculating acres burned per day
different_years_df["acres burned per day"] = different_years_df["Acres Burned"] / different_years_df["days_of_wildfire"]

# Calculating acres burned in the fire started year
different_years_df["acres burned start year"] = different_years_df["acres burned per day"] * different_years_df["Active Days Until Year End"]

different_years_df["acres burned end year"] = different_years_df["Acres Burned"] - different_years_df["acres burned start year"]

different_years_df
```

```python
# Calculating start year data

start_year = different_years_df.groupby("Fire Started year")["acres burned start year"].sum().round(2).reset_index()
start_year.columns = ["Year","Acres Burned (Start Year)"]
start_year
```

```python
# Calculating end year data

end_year = different_years_df.groupby("Fire Extinguished year")["acres burned end year"].sum().round(2).reset_index()
end_year.columns = ["Year","Acres Burned (End Year)"]
end_year
```

```python
# Merging start year and end year data together

merged_years = pd.merge(start_year, end_year, how = "outer")
merged_years
```

```python
# Merging the different years data with same year data

final_data = pd.merge(acres_burned_per_year, merged_years, how = "outer")
final_data.fillna(0, inplace = True)

final_data["Total"] = final_data["Total Acres Burned"] + final_data["Acres Burned (Start Year)"] + final_data["Acres Burned (End Year)"]

final_data
```
