```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Wildfires_df = pd.read_csv("./SMU_Bootcamp_Project-1/california_wildfire_data_cleaned.csv")
```

```python
Wildfires_df[Wildfires_df["County"]== 'Alpine']['Acres Burned'].sum()
```

```python
Freq_Counties_DF = Wildfires_df[(Wildfires_df['County'] == 'Riverside') | (Wildfires_df['County'] == 'Kern') | (Wildfires_df['County'] == 'San Diego') |
                (Wildfires_df['County'] == 'Fresno') | (Wildfires_df['County'] == 'Los Angeles') ]
Freq_Counties_DF
```

```python
County_Counts = Freq_Counties_DF['County'].value_counts()
County_Counts
```

```python
panda_plot = County_Counts.plot.bar()
panda_plot.set_xlabel("County Name")
panda_plot.set_ylabel("Acres Burned")
panda_plot.set_title("#of Wildfires by County")
```

```python
County_Sum = Freq_Counties_DF.groupby(['County'])['Acres Burned'].sum()
County_Sum
```

```python
panda_plot = County_Sum.plot.bar()
panda_plot.set_xlabel("County Name")
panda_plot.set_ylabel("Acres Burned")
panda_plot.set_title("Total Acres Burned by County")
```

```python
County_Avg_Burned = Freq_Counties_DF.groupby(['County'])['Acres Burned'].mean()
County_Avg_Burned
```

```python
panda_plot = County_Avg_Burned.plot.bar()
panda_plot.set_xlabel("County Name")
panda_plot.set_ylabel("Acres Burned")
panda_plot.set_title("Average Acres Burned by County")
```

```python
County_Ppl = Freq_Counties_DF.groupby(['County'])['Personnel Involved'].mean()
County_Ppl
```

```python
panda_plot = County_Ppl.plot.bar()
panda_plot.set_xlabel("County Name")
panda_plot.set_ylabel("Personnel Involved Burned")
panda_plot.set_title("Average Personnel Involved by County")
```
